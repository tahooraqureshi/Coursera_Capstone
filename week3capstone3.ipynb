# Ignoring cells with a borough that is Not assigned

```python
tables=pd.read_csv('tables.csv')
df=pd.DataFrame(tables, columns=['Postcode','Borough', 'Neighbourhood'])
df.replace('Not assigned', np.nan, inplace=True)
df.dropna(subset=['Borough'],how='all',inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

# Merging rows with neighbourhoods seperated by comma

```python
group_df=df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(list).reset_index()
group_df['Neighbourhood']=group_df['Neighbourhood'].apply(lambda x: ','.join(map(str, x)))
group_df
```

# Making Not assigned neighborhood same as the borough

```python
group_df['Neighbourhood'].replace('nan', "Queen's Park",inplace=True)
group_df
```

# Checking shape of dataframe

```python
print("shape:", group_df.shape)
```

# Reading geospatial coordinates file into dataframe

```python
df1=pd.read_csv('Geospatial_Coordinates.csv')
df1.columns=['Postcode', 'Latitude', 'Longitude']
df1
```

# Merging two dataframes to create desired dataframe

```python
result = pd.merge(group_df,
                df1[['Postcode', 'Latitude', 'Longitude']],
                on='Postcode',left_index=True, right_index=True, how='outer')
result
```

# Creating dataframe with borough = Toronto

```python
toronto=result.loc[result['Borough'].str.endswith("Toronto")]
toronto
```

# Visualizing clustered neighbourhoods in the city of Toronto

```python
m = folium.Map([43.676357,-79.293031], zoom_start=11)
for index, row in toronto.iterrows():
    folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=5,
                        popup=row['Neighbourhood'],
                        fill=True,
                        fill_color="#3db7e4", # divvy color
                       ).add_to(m)
m
```
