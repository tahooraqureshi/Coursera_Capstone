# Ignoring cells with a borough that is Not assigned

```python
tables=pd.read_csv('tables.csv')
df=pd.DataFrame(tables, columns=['Postcode','Borough', 'Neighbourhood'])
df.replace('Not assigned', np.nan, inplace=True)
df.dropna(subset=['Borough'],how='all',inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

# Merging rows with neighbourhoods seperated by comma

```python
group_df=df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(list).reset_index()
group_df['Neighbourhood']=group_df['Neighbourhood'].apply(lambda x: ','.join(map(str, x)))
group_df
```

# Making Not assigned neighborhood same as the borough

```python
group_df['Neighbourhood'].replace('nan', "Queen's Park",inplace=True)
group_df
```
